```
import pandas as pd
from pykrx import stock
from datetime import datetime, timedelta

def calculate_index_returns(index_code):
    today = datetime.today().strftime("%Y%m%d")
    end_date = today
    start_date = (datetime.today() - timedelta(days=150)).strftime("%Y%m%d")

    df = stock.get_index_ohlcv(start_date, end_date, index_code)

    # Create a new DataFrame to store returns
    returns_df = pd.DataFrame()

    returns_df['인덱스코드'] = [index_code]
    returns_df['인덱스명'] = [stock.get_index_ticker_name(index_code)]
    # # Get index name and truncate to 7 characters
    # index_name = stock.get_index_ticker_name(index_code)
    # returns_df['인덱스명'] = [index_name[:8]]  # Truncate to 7 characters

    # print(returns_df['인덱스명'].iloc[0], returns_df['인덱스코드'].iloc[0])

    # Calculate and assign returns to the new DataFrame
    returns_df['1d'] = [(df['종가'].pct_change(periods=1) * 100).round(2).iloc[-1]]
    returns_df['3d'] = [(df['종가'].pct_change(periods=3) * 100).round(2).iloc[-1]]
    returns_df['5d'] = [(df['종가'].pct_change(periods=5) * 100).round(2).iloc[-1]]
    returns_df['10d'] = [(df['종가'].pct_change(periods=10) * 100).round(2).iloc[-1]]
    returns_df['20d'] = [(df['종가'].pct_change(periods=20) * 100).round(2).iloc[-1]]
    returns_df['30d'] = [(df['종가'].pct_change(periods=30) * 100).round(2).iloc[-1]]
    returns_df['40d'] = [(df['종가'].pct_change(periods=40) * 100).round(2).iloc[-1]]
    returns_df['50d'] = [(df['종가'].pct_change(periods=50) * 100).round(2).iloc[-1]]
    returns_df['60d'] = [(df['종가'].pct_change(periods=60) * 100).round(2).iloc[-1]]
    returns_df['90d'] = [(df['종가'].pct_change(periods=90) * 100).round(2).iloc[-1]]
    # returns_df['120d'] = [(df['종가'].pct_change(periods=120) * 100).round(2).iloc[-1]]

    # Select only the last row using tail(1)
    returns_df = returns_df.tail(1)

    return returns_df

if __name__ == "__main__":
    index_codes = ['1001', '2001', '1002', '1003', '1004', '2002', '2003', '2004']  # Example ETF codes

     # Iterate through each index code and call calculate_index_returns
    results = []
    for code in index_codes:  # Iterate through each code
        result = calculate_index_returns(code) # Pass a single code to calculate_index_returns
        results.append(result)

    # 결과를 하나의 DataFrame으로 합치기
    all_returns_df = pd.concat(results)

    all_returns_df = all_returns_df.set_index(['인덱스코드', '인덱스명'])

    # # Apply styling to format index values
    # styled_df = all_returns_df.style.set_properties(
    #     subset=pd.IndexSlice[:, :],  # Apply to both index levels
    #     **{'width': '20px', 'text-align': 'left'}  # Set width and alignment
    # )

    # Display the styled DataFrame in Jupyter Notebook
    # display(styled_df)


    print(all_returns_df)
    # all_returns_df.to_excel("ETF_수익률분석.xlsx")
```

```
#당일 수익률
import pandas as pd
from pykrx import stock
from datetime import datetime, timedelta
import concurrent.futures
import time

def calculate_etf_returns(etf_code):
    today = datetime.today().strftime("%Y%m%d")
    end_date = today
    start_date = (datetime.today() - timedelta(days=2)).strftime("%Y%m%d")

    df = stock.get_etf_ohlcv_by_date(start_date, end_date, etf_code)
    # print(df)
    time.sleep(3)

    if df.empty or not all(col in df.columns for col in ['종가', '시가', '거래대금']):
        print(f"Error: Data not found or missing '종가' column for ETF code: {etf_code}")
        return pd.DataFrame()


    # Create a new DataFrame to store returns
    returns_df = pd.DataFrame()

    returns_df['종목코드'] = [etf_code]
    returns_df['종목명'] = [stock.get_etf_ticker_name(etf_code)]

    print(returns_df['종목명'].iloc[0], returns_df['종목코드'].iloc[0])

    # Calculate and assign returns to the new DataFrame
    returns_df['1d'] = [(df['종가'].pct_change(periods=1) * 100).round(2).iloc[-1]]

    if df['시가'].iloc[-1] == 0:  # Check if '시가' is 0
        returns_df['시가대비'] = 0  # Assign 0 if '시가' is 0
    else:
        returns_df['시가대비'] = round((df['종가'].iloc[-1] / df['시가'].iloc[-1] - 1) * 100, 2)
    # returns_df['시가대비'] = round((df['종가'].iloc[-1] / df['시가'].iloc[-1] - 1) * 100, 2)
    returns_df['거래대금'] = [df['거래대금'].iloc[-1]]
    returns_df['거래대금'] = returns_df['거래대금'].apply(lambda x: "{:,.0f}억".format(x / 100000000))

    # Select only the last row using tail(1)
    # returns_df = returns_df.tail(1)

    return returns_df

if __name__ == "__main__":
    etf_codes = ['069500', '498400', '122630', '252670', '371460']  # Example ETF codes

    code_list = pd.read_excel("/content/ETF분류.xlsx", sheet_name="코드")
    # print(code_list)

    # Extract ETF codes from the '종목코드' column
    etf_codes = code_list['종목코드'].tolist()
    # print(etf_codes)

    with concurrent.futures.ThreadPoolExecutor() as executor:
        # 각 ETF 코드에 대해 calculate_etf_returns 함수를 비동기적으로 실행
        results = list(executor.map(calculate_etf_returns, etf_codes))

    # 결과를 하나의 DataFrame으로 합치기
    all_returns_df = pd.concat(results)

    # Calculate returns for each ETF and combine into a single DataFrame
    # all_returns_df = pd.concat([calculate_etf_returns(code) for code in etf_codes])

    # Set '종목코드' and '종목명' as index
    all_returns_df = all_returns_df.set_index(['종목코드', '종목명'])
    all_returns_df = all_returns_df.sort_values(by=['1d'], ascending=False)  # Sort by '1d' column
    print(all_returns_df.head(50))
    # all_returns_df.to_excel("ETF_수익률분석.xlsx")
```
